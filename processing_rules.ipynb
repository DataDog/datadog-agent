```python
import numpy as np
import pandas as pd
import ollama
import json
```

```python
data_path = "/Users/ella.taira/Downloads/logs_config_processing_rules_2025-12-08T22_06_45.272468915Z.csv" 
df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
df = df.drop(columns=["ORG_ID", "Agent count (avg over time period)"])
```

```python
df["config_value"] = df["Config value"]
```

```python
df = df.drop(columns=["Config value"])
```

```python
# Parse JSON and explode so each rule gets its own row
df["data"] = df["config_value"].apply(json.loads)
df = df.explode("data", ignore_index=True)

# Expand each rule dict into columns
df = df.join(df["data"].apply(pd.Series)).drop(columns=["data"])

print(df.head())
print(f"\nTotal rows: {len(df)}")
print(f"\nColumns: {df.columns.tolist()}")
```
